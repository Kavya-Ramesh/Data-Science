```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_excel('C:\\Project -DS\\DEXINUS (1).xls') #skiprows = range(0,10)
data
```

```python
data = data.rename({'observation_date':'date','DEXINUS':'rate'},axis=1)
```

```python
data.ffill(inplace=True)
```

```python
data.isnull().sum()
```

```python
# extract month and year from dates
data['month'] = pd.DatetimeIndex(data['date']).month
data['year'] =  pd.DatetimeIndex(data['date']).year

#data1["month"] = data1.Date.dt.strftime("%b") # month extraction
#data1["year"] = data1.Date.dt.strftime("%y") # year extraction

# create a sequence of numbers
data['Series'] = np.arange(1,len(data)+1)# drop unnecessary columns and re-arrange 
data = data[['Series', 'year', 'month', 'rate']] 

# check the head of the dataset
data.head()
```

```python
data.tail()
```

```python
# split data into train-test set
train = data[data['year'] < 2019]
test = data[data['year'] >= 2019]
# check shape
train.shape, test.shape
```

```python
# import the regression module
from pycaret.regression import *
# initialize setup
s = setup(data = train , test_data = test, target = 'rate',
          fold_strategy = 'timeseries', numeric_features = ['year', 'Series'],
          fold = 3, transform_target = True, session_id = 123)
```

```python
best = compare_models(sort = 'MAE')
```
