```python
crimek = pd.read_csv('C:\\Data Science\\Clustering\\crime_data.csv')
crimek
 
```

```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn.cluster import KMeans
```

```python
 from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crimek_df = scaler.fit_transform(crimek.iloc[:,1:])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crimek_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
 
clusters_new = KMeans(5, random_state=70)
clusters_new.fit(scaled_crimek_df)
```

```python
clusters_new.labels_
```

```python
crimek['clusterid_new'] = clusters_new.labels_
crimek['clusterid_new'].value_counts()
```

```python
crimek['clusterid_new'] = clusters_new.labels_
crimek
```

```python
crimek['Total_crime'] = crimek['Murder'] + crimek['Assault'] + crimek['UrbanPop'] + crimek['Rape']
```

```python
clusters_new.cluster_centers_
```

```python
crimek.groupby('clusterid_new').agg(['mean']).reset_index()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbcrime =pd.read_csv('C:\\Data Science\\Clustering\\crime_data.csv')
dbcrime.head()
```

```python
dbcrime.info()
```

```python
dbcrime.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
array = dbcrime.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN (eps=1 , min_samples=3)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
cl.value_counts()
```

```python
pd.concat([dbcrime,cl],axis=1)
```

```python
dbcrime['cluster']= dbscan.labels_
dbcrime
```
